```python
def log_2_comb(n,k):
    res = 0.0
    for i in range(k):
        res += math.log2(n - i) - math.log2(i + 1)
    return k + res
log_2_comb(784, 16)
```

```python
x = np.linspace(0, math.sqrt((32*32*3) / 6 - 7/120) + 3*math.sqrt(7/120), 1000)

fig, ax = plt.subplots(1)
ax.plot(x, stats.norm.pdf(x, math.sqrt((784) / 6 - 7/120), math.sqrt(7/120)), label=f"MNIST (image, 784 dim)")
ax.plot(x, stats.norm.pdf(x, math.sqrt((2381) / 6 - 7/120), math.sqrt(7/120)), label=f"EMBER (malware, 2381 dim)")
ax.plot(x, stats.norm.pdf(x, math.sqrt((32*32*3) / 6 - 7/120), math.sqrt(7/120)), label=f"CIFAR (image, {32*32*3} dim)")
ax.set_xlabel("Distance")
ax.set_ylabel(f'# of random pairs with that distance')
ax.legend()
fig.set_size_inches(8, 4)
fig.set_dpi(500)
plt.show()
fig.savefig("images/dist_distribution")
```

```python
def mean_dist_to_closest(dim, n, k):
    term1 = (1/dim)*math.lgamma(dim/2 + 1) - math.log(math.sqrt(math.pi)) 
    term2 = math.lgamma(k + 1/dim) - math.lgamma(k)
    term3 = math.lgamma(n) - math.lgamma(n + 1/dim)
    print(term1, term2, term3)
    return math.exp(term1 + term2 + term3)
mean_dist_to_closest(784,50_000,1)
```
